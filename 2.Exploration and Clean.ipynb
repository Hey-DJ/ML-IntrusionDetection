### Data quality checks


```python
#Check if any of the columns have any missing/NaN data
colNulls = events.isna().sum()
print('Columns with NaN values:', (colNulls[ colNulls != 0 ]).count())
if ((colNulls[ colNulls != 0 ]).count() > 0):
    print('Columns: ', colNulls[ colNulls != 0 ])
```

```python
#What is the value of the Flow Pkts/s feature (which is similar to the Flow Byts/s feature)
events[ events['Flow Byts/s'].isnull()]['Flow Pkts/s'].value_counts()
```

```python
#What is the value of the Flow Duration feature (as Flow Byts/s metric is calculated using the Flow Duration feature - eg Flow Bytes/s = Bytes in Flow / Flow Duration)
events[ events['Flow Byts/s'].isnull()]['Flow Duration'].value_counts()
```

```python
#How many samples are affected by this Flow Duration=0 issue?
events[ events['Flow Duration']==0]['Flow Byts/s'].value_counts()
```

```python
#How many samples are affected by this Flow Duration=0 issue?
events[ events['Flow Duration']==0]['Label'].value_counts()
```

**Observation:** The above analysis shows that the root cause of the 59215 samples having NaN in the 'Flow Pkts/s' features is due to the Flow Duration feature being set to 0 for whatever reason (which then in turn causes division-by-zero error when other features are being calculated). Given the Flow Duration feature is measured in microseconds, we expect every sample should have a valid value here. Some error would have occured during the network packet capture or during the generation of flow related features.
As this issue has also resulted in NaN and INF values in the 'Flow Byts/s' and 'Flow Pkts/s' features, it would make sense to completely remove these samples altogether (even before we get to the stage of allocating training vs testing sets). The impact of the removal would also be small as there are only 93836 samples afflicted by this issue

```python
events.drop(axis=1, index=events[ events['Flow Duration']==0].index, inplace=True)
```

### Delete invalid data

During the evalution stage of a suitable classifier, it became apparent from analysis on the Confusion Matrix, that some of the 'FTP-BruteForce' samples were being incorrectly predicted as 'DoS attacks-SlowHTTPTest', while large number of the 'DoS attacks-SlowHTTPTest' samples were being incorrectly predicted as 'FTP-BruteForce'. Further inspection of these samples revealed that they are almost identical. 

```python
from CustomFunctions import printLarge

printLarge(events[ events['Label']=='DoS attacks-SlowHTTPTest'])
```

```python
from CustomFunctions import printLarge

printLarge(events[ events['Label']=='FTP-BruteForce'])
```

**Observation:** Note how nearly all the features look almost identical. Even these highly variable features show similarities: [Init Fwd Win Bytes], [Fwd Seg Size Min], [Fwd Pkts/s], [Fwd Header Len]. What is also surprising is that the DST Port feature is "21" for both of these classes. Port 21 is expected for FTP-BruteForce attacks (as port 21 is a well known port for FTP). However, for 'DoS attacks-SlowHTTPTest', the expected port is either 80 (for HTTP attacks) or port 443 (for HTTPS attacks). So it is highly likely that the 'DoS attacks-SlowHTTPTest' samples were created with the incorrect label. Most likely those samples are actually FTP-Bruteforce attacks, but incorrectly labelled as 'DoS attacks-SlowHTTPTest'.

Hence the decision was made to completely delete these 'DoS attacks-SlowHTTPTest' samples.

```python
print('Dataset size before deleting DoS attacks-SlowHTTPTest samples: ', events.shape)
print('DoS attacks-SlowHTTPTest samples to be deleted: ', len(events[ events['Label']=='DoS attacks-SlowHTTPTest']))
events.drop(axis=0, index=events[ events['Label']=='DoS attacks-SlowHTTPTest'].index, inplace=True)
print('Dataset size after deleting DoS attacks-SlowHTTPTest samples: ', events.shape)
```
